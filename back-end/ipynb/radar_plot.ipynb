```python
import matplotlib.pyplot as plt
import numpy as np

# Updated function with color customization
def create_comparative_radar_chart_colored(ax, categories, values_dict, title, colors):
    N = len(categories)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]  # complete the circle

    for (label, values), color in zip(values_dict.items(), colors):
        values = values + values[:1]
        ax.plot(angles, values, label=label, marker='o', color=color)
        ax.fill(angles, values, alpha=0.3, color=color)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=14)
    ax.set_yticks([1, 2, 3, 4, 5])
    ax.set_yticklabels(['1', '2', '3', '4', '5'], fontsize=14)
    ax.set_ylim(0, 5)
    ax.set_title(title, size=14, pad=20)
    ax.legend(loc='upper right', bbox_to_anchor=(1.6, 0.1), fontsize=14)

# Data
doc_categories = ['Relevance', 'Accuracy', 'Usefulness', 'Temporality', 'Actionability']
doc_models = {
    'all-MiniLM-L6-v2': [3.78, 4.45, 3.66, 4.45, 2.57],
    'Qwen3-Embedding-0.6B': [4.26, 4.99, 3.89, 4.99, 2.08]
}
doc_colors = ['#FEA715', '#9895F9']

ans_categories = ['Congruence', 'Coherence', 'Relevance', 'Creativity', 'Engagement']
ans_models = {
    'Qwen3-0.6B': [2.27, 3.29, 2.44, 1.94, 2.06],
    'Qwen3-1.7B': [3.41, 3.94, 4.18, 3.39, 2.51]
}
# Use light and deep purple for answer generation models
ans_colors_purple = ['#9895F9', '#6B29EC']  # Qwen3-0.6B = lighter, Qwen3-1.7B = deeper

# Re-plotting with updated colors
fig, axes = plt.subplots(1, 2, figsize=(14, 6), subplot_kw=dict(polar=True))

create_comparative_radar_chart_colored(axes[0], doc_categories, doc_models, "Document Retriever Comparison", doc_colors)
create_comparative_radar_chart_colored(axes[1], ans_categories, ans_models, "Answer Generator Comparison", ans_colors_purple)

plt.tight_layout()
plt.show()
```
